```python
import rasterio
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
import seaborn as sns
def plot_hist(data):
    sns.displot(
   data=data, kind='kde',bw_adjust=.25
)
```

```python
ep4=rasterio.open(f"Test_output_ep_v2_4.tif").read() #these files must be generated before this step
ep5=rasterio.open(f"Test_output_ep_v2_5.tif").read()
ep6=rasterio.open(f"Test_output_ep_v2_6.tif").read()
ep7=rasterio.open(f"Test_output_ep_v2_7.tif").read()
ep8=rasterio.open(f"Test_output_ep_v2_8.tif").read()
ep9=rasterio.open(f"Test_output_ep_v2_9.tif").read()
ep10=rasterio.open(f"Test_output_ep_v2_10.tif").read()
ep11=rasterio.open(f"Test_output_ep_v2_11.tif").read()
```

```python
df_sus = pd.DataFrame(data=np.stack([ep4[0].flatten(),ep5[0].flatten(),ep6[0].flatten(),ep7[0].flatten(),ep8[0].flatten(),ep9[0].flatten(),ep10[0].flatten(),ep11[0].flatten()]).swapaxes(0,1), columns=["Epoch 4", "Epoch 5","Epoch 6","Epoch 7", "Epoch 8","Epoch 9", "Epoch 10","Epoch 11"])
```

```python
df_int = pd.DataFrame(data=np.stack([ep4[1].flatten(),ep5[1].flatten(),ep6[1].flatten(),ep7[1].flatten(),ep8[1].flatten(),ep9[1].flatten(),ep10[1].flatten(),ep11[1].flatten()]).swapaxes(0,1), columns=["Epoch 4", "Epoch 5","Epoch 6","Epoch 7", "Epoch 8","Epoch 9", "Epoch 10","Epoch 11"])
```

```python
df_haz = pd.DataFrame(data=np.stack([ep4[2].flatten(),ep5[2].flatten(),ep6[2].flatten(),ep7[2].flatten(),ep8[2].flatten(),ep9[2].flatten(),ep10[2].flatten(),ep11[2].flatten()]).swapaxes(0,1), columns=["Epoch 4", "Epoch 5","Epoch 6","Epoch 7", "Epoch 8","Epoch 9", "Epoch 10","Epoch 11"])
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(figsize=(20,20), ncols=2, nrows=4,constrained_layout=True)
sns.kdeplot(data=df_sus["Epoch 4"],ax=axes[0,0],fill=True).set(xlim=(0,1.2),ylim=(0,20),ylabel='Kernel Density',xlabel='Susceptibility',title='Epoch 4')
sns.kdeplot(data=df_sus["Epoch 5"],ax=axes[0,1],fill=True).set(xlim=(0,1.2),ylim=(0,20),ylabel='Kernel Density',xlabel='Susceptibility',title='Epoch 5')
sns.kdeplot(data=df_sus["Epoch 6"],ax=axes[1,0],fill=True).set(xlim=(0,1.2),ylim=(0,20),ylabel='Kernel Density',xlabel='Susceptibility',title='Epoch 6')
sns.kdeplot(data=df_sus["Epoch 7"],ax=axes[1,1],fill=True).set(xlim=(0,1.2),ylim=(0,20),ylabel='Kernel Density',xlabel='Susceptibility',title='Epoch 7')
sns.kdeplot(data=df_sus["Epoch 8"],ax=axes[2,0],fill=True).set(xlim=(0,1.2),ylim=(0,20),ylabel='Kernel Density',xlabel='Susceptibility',title='Epoch 8')
sns.kdeplot(data=df_sus["Epoch 9"],ax=axes[2,1],fill=True).set(xlim=(0,1.2),ylim=(0,20),ylabel='Kernel Density',xlabel='Susceptibility',title='Epoch 9')
sns.kdeplot(data=df_sus["Epoch 10"],ax=axes[3,0],fill=True).set(xlim=(0,1.2),ylim=(0,20),ylabel='Kernel Density',xlabel='Susceptibility',title='Epoch 10')
sns.kdeplot(data=df_sus["Epoch 11"],ax=axes[3,1],fill=True).set(xlim=(0,1.2),ylim=(0,20),ylabel='Kernel Density',xlabel='Susceptibility',title='Epoch 11')
fig.suptitle('Susceptibility Over Time', fontsize=24)
#sns.rugplot(data=df_sus["Epoch 4"],ax=axes[0,0])
```

```python
fig.savefig("Susceptibility_over_time_v2.png",dpi=500)
```
